```python
MACs = []
for i in range(1, 13):
    pre = "00:" * 5
    MACs.append(pre + f'{i:02}')
subnet_ips = []
server_ips = []
switch_ips = []
for i in range(1, 4):
    subnet = f'192.168.{i}.0/24'
    subnet_ips.append(subnet)
    server = f'192.168.{i}.10/24'
    server_ips.append(server)
    switch = f'192.168.{i}.1/24'
    switch_ips.append(switch)
```

```python
i, j = 0, 0

for device in devices:
    device_info = devices[device]
    node = device_info['node']
    if device_info['type'] == 'switch':
        net_name = f'server{i+1}_switch{i+1}_net'
        iface = device_info['interfaces'][net_name]['name']
        ip = switch_ips[i]
        node.execute(f'sudo ifconfig {iface} {ip}')
        print(device + " " + iface + " " + ip)
        topology['devices'][device]['interfaces'][net_name]['IP'] = ip
        i += 1
        
        
    else:
        net_name = f'server{j+1}_switch{j+1}_net'
        iface = device_info['interfaces'][net_name]['name']
        ip = server_ips[j]
        node.execute(f'sudo ifconfig {iface} {ip}')
        print(device + " " + iface + " " + ip)
        topology['devices'][device]['interfaces'][net_name]['IP'] = ip
        j += 1 

pp(topology)
```
